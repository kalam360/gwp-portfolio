```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# 1.1 Obtain returns for 11 SPDRs

etf_list = ['XLC','XLY','XLP','XLE','XLF','XLV','XLI','XLB','XLRE','XLK','XLU']

prices_df = pd.DataFrame()

for etf in etf_list:
    df = yf.download(etf, start='2014-01-02', end='2019-12-31', progress=False)
    df[etf] = df['Adj Close']
    prices_df = pd.concat([prices_df, df[etf]], axis = 1)

prices_df.head(10)
```

```python
# 1.2 Obtain US equity prices, we decided to use the Vanguard Growth ETF (Ticker: VUG)

fund = yf.download('VUG', start='2014-01-02', end='2019-12-31', progress=False)
df['VUG'] = df['Adj Close']
prices_df = pd.concat([prices_df, df['VUG']], axis = 1)

prices_df.head(10)
```

```python
# Computing daily returns for all funds

etf_list.append(['VUG'])

returns_df = pd.DataFrame()

for etf in etf_list:
    returns = np.log(prices_df[etf]/prices_df[etf].shift(1)).dropna()
    returns_df = pd.concat([returns_df, returns], axis = 1)

returns_df.head(5)
```

```python
# 3.1 Computing means

print("Mean returns:")
mean = returns_df.mean(axis = 0)
print(mean)

lower_bound = returns_df.quantile(q=0.05, axis=0)
upper_bound = returns_df.quantile(q=0.95, axis=0)
returns_trimmed_5pct = returns_df.clip(lower=lower_bound,upper=upper_bound,axis=1)
returns_trimmed_mean = returns_trimmed_5pct.mean(axis=0)

print ("")
print("5% trimmed means:")
print(returns_trimmed_mean)

plt.scatter(returns_df.mean(axis = 0).index, returns_df.mean(axis = 0), label = 'Mean Returns')
plt.scatter(returns_trimmed_mean.index, returns_trimmed_mean, label = 'Trimmed Returns')
plt.xlabel("ETF")
plt.ylabel("Return")
plt.legend()
print("")
print("The returns for XLE seem to be the outlier, it is the only negative return")
```

```python
# 3.2 Interquartile range

print("**Interquantile Range:**")
print("")
print("1st quantile:")
print(returns_df.quantile(0.25, axis = 0))
print("2nd quantile:")
print(returns_df.quantile(0.5, axis = 0))
print("3rd quantile:")
print(returns_df.quantile(0.75, axis = 0))

print("")
print("This is a plot of the quantiles")
returns_df.boxplot(figsize = (12, 6))
plt.show()

print("")
print("Standard deviation: ")
print(returns_df.std(axis = 0))

print("")
print("The larger the standard deviation of the returns, the wider the interquantile range will be.")
```

```python
# 3.3 Mean, Median, Skew

print("Median Returns:")
median = returns_df.median(axis = 0)
print(median)

plt.scatter(mean.index, mean, label = "Mean")
plt.scatter(median.index, median, label = "Median")
plt.legend()
plt.show()

skew = returns_df.skew(axis = 0)
print("Skew:")
print(skew)

print("")
print("All the funds have a negative skew, that is why the median returns are higher than the mean returns.")
```

```python
# 3.4 Kurtosis

kurt = returns_df.kurtosis(axis = 0)
print("Kurtosis:")
print(kurt)
```

```python
# 4.1 Data Inference - Fitting to Normal Distribution

print("We will use XLY as the first ETF and fit to both a normal distribution and a heavy-tail distribution.")
print("We can see that the normal distribution does not fit the data well.")

import seaborn as sns
from scipy.stats import norm

sns.distplot(returns_df['XLY'], label = "Heavy tail")
sns.distplot(returns_df['XLY'], fit = norm, kde = False, label = "Normal")
plt.title("Returns of XLY fitted to normal distribution")
plt.legend()
plt.show()
```

```python
# 4.2 Fitting another ETF

print("We will use XLK as the second ETF and fit to both a normal distribution and a heavy tail distribution.")
print("We can see again that the normal distribution does not fit the data well.")

import seaborn as sns
from scipy.stats import norm

sns.distplot(returns_df['XLK'], label = "Heavy tail")
sns.distplot(returns_df['XLK'], fit = norm, kde = False, label = "Normal")
plt.title("Returns of XLK fitted to normal distribution")
plt.legend()
plt.show()
```
